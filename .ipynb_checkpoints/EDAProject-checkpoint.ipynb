```python
# import all libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
from dateutil import parser
#import parser as ps
```

```python
df1=pd.read_csv('appdata10.csv')
df1.hour=df1.hour.str.slice(1,3).astype(int)
```

```python
df1.head()
df1.describe()
df1.info()
print(df1.columns)
print(df1.hour.head())
```

```python
# create a new df2 with partial columns
df2=df1.copy().drop(columns=['user','screen_list','enrolled_date','first_open','enrolled'])
df2.head()
```

```python
# plot histograms

plt.suptitle('Histograms of Numberical Columns', fontsize=10)
for i in range(1, df2.shape[1]+1):
    plt.subplot(3,3,i)
    f=plt.gca()
    f.set_title(df2.columns.values[i-1])
    vals=np.size(df2.iloc[:,i-1].unique())
    plt.hist(df2.iloc[:,i-1], bins=vals, color = '#3F5D7D')
```

```python
# Find the correlation of all df2 columns with one of the DF1 column.  df1 column is called response variable
# Bigger the correlation value (+/-) more impact to the enrollment.
print(df2.corrwith(df1.enrolled))
df2.corrwith(df1.enrolled).plot.bar(figsize =(20,10), title="Correlation with Response Variable",
                                   fontsize = 15, rot=45, grid=True)
```

```python
# Find the Corrletion Matrix

sn.set(style="white", font_scale=2)   # define a graph

corr=df2.corr()                       # define correlation of all columns in df2 NxN (7x7) matrix

mask = np.zeros_like(corr, dtype=np.bool)   # mask is a boolean matirx of 7x7. same shape as corr
mask[np.triu_indices_from(mask)]=True  # triu is the upper triangle.  This set all values in the upper tri to true
                                        # meaning those cell with True will be omitted

f, ax=plt.subplots(figsize=(18,10))   # Create a subplot with 1 plot for better graphic features
f.suptitle("Correlation Matrix", fontsize=40)  # Set the title

cmap=sn.diverging_palette(220,10, as_cmap=True) # create diverging colormap

sn.heatmap(corr, mask=mask, cmap=cmap, vmax=0.3, center=0, 
           square=True, linewidths=0.5, cbar_kws={"shrink": 0.5})
```

```python
# Make a copy of df so df1 will not be changed.  
# If something happened to df3, it can be rerun from here
df3=df1.copy()
#print(df3.dtypes)
print(df3.info())
print(df3.describe())
```

```python
# parser will parse string '2012-12-27 02:14:51.273' into timestamp value '(2012, 12, 27, 2, 14, 51, 273000)'

df3["first_open"]=[parser.parse(row_data) for row_data in df1["first_open"]] 
df3["enrolled_date"]=[parser.parse(row_data) if isinstance(row_data, str) else row_data for row_data in df1["enrolled_date"] ]
df3["difference"]=(df3.enrolled_date-df3.first_open).astype('timedelta64[h]')
plt.hist(df3['difference'].dropna(), color='#3F5D7D')
plt.title('Time Since First Open to Enrolled')
plt.show()

plt.hist(df3['difference'].dropna(), color='#3F5D7D', range=[0,100])
plt.title('Time Since First Open to Enrolled')
plt.show()

df3.loc[df3.difference > 48, 'enrolled'] = 0   # set all those enrolled over 48 hours as not enrolled.
df3=df3.drop(columns=['difference', 'enrolled_date','first_open'])
```

```python
## FOrmatting the screen_list Field

top_screens=pd.read_csv('top_screens.csv').top_screens.values  # This assign N rows to one rows of N values
#top_screens=pd.read_csv('top_screens.csv')
df3["screen_list"] = df1.screen_list.astype(str) + ','        # Add a ',' to the end of screenlist  
for sc in top_screens:                                        # THIS LOOP CANNOT BE RUN 2ND TIME
    df3[sc] = df3.screen_list.str.contains(sc).astype(int)    # This will create a new column with value 1 if sc 
                                                              # is in the screen_list.
    df3["screen_list"] = df3.screen_list.str.replace(sc+",", "")  # This will replace the top screen name with ""
    
```

```python
df3["Other"]=df3.screen_list.str.count(",")
df3 = df3.drop(columns=(["screen_list"]))
```

```python
savings_screens=["Saving1",
                "Saving2",
                "Saving2Amount",
                "Saving4",
                "Saving5",
                "Saving6",
                "Saving7",
                "Saving8",
                "Saving9",
                "Saving10",
               ]
df3["SavingsCount"]=df3[savings_screens].sum(axis=1)
#df3["SavingsCount"]
```

```python
credit_screens=["Credit1",
                "Credit2",
                "Credit3",
                "Credit3Container",
                "Credit3Dashboard"
                ]
df3["CreditCount"]=df3[credit_screens].sum(axis=1)
#df3["CreditCount"]
```

```python
cc_screens=["CC1",
            "CC1Category",
            "CC3"]
df3["CCCount"]=df3[cc_screens].sum(axis=1)
#df3["CCCount"]
```

```python
loan_screens=["Loan",
              "Loan2",
              "Loan3",
              "Loan4"]
        
df3["LoanCount"]=df3[loan_screens].sum(axis=1)
#df3["CreditCount"]
```

```python
# Drop the actual screen columns since we already have the column count
df3 = df3.drop(columns = (savings_screens + credit_screens + cc_screens+ loan_screens))
```

```python
# Finally we want to save the file to the new file call new_appdata10.csv
df3.to_csv("../data/new_appdata10.csv", index =False)
```
