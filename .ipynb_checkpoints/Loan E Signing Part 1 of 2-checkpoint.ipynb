This is a loan application AI model to determine if an applicant is likely to complete the e-signing process or not.
It is not to determine if the applicant will get the loan or not.
Part 1 will preprocess the data, create different charts to study the columns and remove the NULL columns and combine some columns

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
    
```

```python
# read the data
df1=pd.read_csv('P39-Financial-Data.csv')  # samples of previous applicant data
# To combine all years_employed and months_employed
df1['total_months_employed'] = df1.months_employed+df1.years_employed*12
df1['personal_account_t'] = df1.personal_account_m + df1.personal_account_y*12
# To drop the origin months_employed, years_employed, personal_account_m, personal_account_y
df1=df1.drop(columns=['months_employed','years_employed','personal_account_m','personal_account_y'])
df2=df1.drop(columns=['entry_id','e_signed'])

```

```python
# check if any null values.  If so, drop the high null columns

#print(df1.isna().any())                # check if any na entries
#print(df1.isna().sum())              # find total na sum 
df1.isna().sum()

# Since there is no null, the following are for reference only
#df1=df1[pd.notnull(df1['age'])]     # Will get only Age not null.  
#df1=df1.drop(columns = ['credit_score', 'rewards_earned'])  # drop two columns becuase of high null occurence
```

```python
# create a histogram for each column


fig=plt.figure(figsize=(15, 12))
plt.suptitle('Histograms of Numberical Columns', fontsize=20)
print(df2.dtypes)
for i in range(1, df2.shape[1]+1):
    if df2.dtypes[i-1] != object:              # problem with object type, so skip the object
        plt.subplot(6, 5, i)
        f=plt.gca()
        f.axes.get_yaxis().set_visible(False)
        f.set_title(df2.columns.values[i-1])
        vals=np.size(df2.iloc[:,i-1].unique())       # Vals is not use because it takes too long. Need more cpu ?
        plt.hist(df2.iloc[:, i - 1],  histtype='bar',color='#3F5D7D')
   #     plt.hist(df2.iloc[:, i - 1], bins=vals, color='#3F5D7D')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
```

```python
# To display correlation to 'e_signed'.  This can be use to pick only highly correlated columns
df2.corrwith(df1.e_signed).plot.bar(
    figsize=(20,10), title="Corrleation with Response Variable", fontsize=10, rot=45,
    grid=True)
```

```python
# Correlation Matrix
sn.set(style='white')

#Compute the correlation matrix
corr=df2.corr()

# Generate a mask for the upper triangle
mask= np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)]=True

# Set up the matplotlib figure
f, ax=plt.subplots(figsize=(18,15))

# Generate a customer diverging colormap
cmap = sn.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sn.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
          square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
# save df1
df1.to_csv("new-Financial-Data.csv")
```
